```python

# writing custom mean function that returns mean
def gpa_mean_func(df):
    
    
    weighted_gpas = df * np.array(
        [4, 3.5, 3, 2.5, 2, 1.5, 1, 0])

    
        
    final_means = np.sum(weighted_gpas)
            
   # return mean

    return final_means
        
        
        
           
custom_mean = probability_grouped.apply(gpa_mean_func)
```

```python
def gpa_std(df):
    mean_group_values = gpa_mean_func(df)
    
    gpa_real_values = np.array(
        [4, 3.5, 3, 2.5, 2, 1.5, 1, 0])
    
    
    x_minus_mean_list = []
    
    df = np.array(df)
    
    for i in range(8):
        
        prob_scaled = df[i] * (
            gpa_real_values[i] - 
            mean_group_values) ** 2  


        x_minus_mean_list.append(prob_scaled)
    
    variance = np.sum(x_minus_mean_list)
    
    std = np.sqrt(variance)
    return std

custom_std = probability_grouped.apply(gpa_std)
```

```python
def gpa_skew(df):
    mean_group_values = gpa_mean_func(df)
    
    gpa_real_values = np.array(
        [4, 3.5, 3, 2.5, 2, 1.5, 1, 0])
    
    
    x_minus_mean_list = []
    
    df = np.array(df)
    
    for i in range(8):
        
        prob_scaled = df[i] * (
            gpa_real_values[i] - 
            mean_group_values) ** 3  


        x_minus_mean_list.append(prob_scaled)
    
    skew_pre  = np.sum(x_minus_mean_list)
    
    skew_post = skew_pre / gpa_std(df) ** 3
    
    return skew_post

    

custom_skew = probability_grouped.apply(gpa_skew)
```

```python
def gpa_kurtosis(df):
    mean_group_values = gpa_mean_func(df)
    
    gpa_real_values = np.array(
        [4, 3.5, 3, 2.5, 2, 1.5, 1, 0])
    
    
    x_minus_mean_list = []
    
    df = np.array(df)
    
    for i in range(8):
        
        prob_scaled = df[i] * (
            gpa_real_values[i] - 
            mean_group_values) ** 4  


        x_minus_mean_list.append(prob_scaled)
    
    kurt_pre  = np.sum(x_minus_mean_list)
    
    kurt_post = kurt_pre / gpa_std(df) ** 4
    
    return kurt_post

    

custom_kurtosis = probability_grouped.apply(gpa_kurtosis)
```

```python
def gpa_mode(df):
    
    
    gpa_real_values = np.array(
        [4, 3.5, 3, 2.5, 2, 1.5, 1, 0])
    
    
 
    
    df = np.array(df)
    
    mode = gpa_real_values[np.argmax(df)]
    
    
    
    return mode

custom_mode = probability_grouped.apply(gpa_mode)
```

```python
def gpa_median(df):
    
    
    gpa_real_values = np.array(
        [4, 3.5, 3, 2.5, 2, 1.5, 1, 0])

    
    
    df = np.array(df)
    
    
    prob_sum = 0
        
    prob_counter = 0
    
    
    
    while prob_sum < .5:

        prob_sum = prob_sum + df[prob_counter]

        
        # if prob sum is greater than 5, don't tick up the counter
        
        
        if prob_sum < .5:
            
            prob_counter = prob_counter + 1
            
    median = gpa_real_values[prob_counter]
            

            
          
    
    return median


custom_median = probability_grouped.apply(gpa_median)
```

3. Individual Fitness: an individuals fitness function is weighted based off of the mean, median, mode, as well as standard deviation
