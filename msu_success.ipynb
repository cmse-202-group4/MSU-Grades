## Import packages

```python
#import the package
import pandas as pd
import numpy as np
```

## Reading in data

```python
grades = pd.read_csv("new_readable_grades.csv")
grades[0:3]
```

## Clear the data

```python
grades = grades.drop(['course_code', 'subject_code'], axis = 1)
grades[0:3]
```

## Checking NA Values

```python
grades.isna().sum()#find the null place
```

```python
grades = grades.dropna()#drop the nulls
grades['average_grade']  = round(grades['average_grade'],2)
grades[0:3]
```

```python
#calculate the number of students passing, and the number of students failing the course.
grades['pass'] = grades['4']+ grades['3.5']+ grades['3']+ grades['2.5']+ grades['2']+ grades['1.5']+grades['1']
grades['fail'] = grades['total_grades']-grades['pass']

#calculate percent passing
grades['Passing Rate(%)'] = round(grades['pass']/grades['total_grades']*100,2)

grades[0:3]
```
